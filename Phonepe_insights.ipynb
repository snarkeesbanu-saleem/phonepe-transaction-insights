## **Project Summary**

This project analyzes the **PhonePe Pulse** dataset, which contains real transaction data across India from 2018 to 2024. The objective was to extract, clean, transform, and visualize transaction patterns to derive actionable business insights.

Key activities included:
- Cloning and processing raw JSON data from PhonePe Pulse GitHub repository
- Creating a clean `aggregated_transaction.csv` with 5000+ records
- Performing comprehensive EDA with **15 different chart types**
- Building an interactive Streamlit dashboard
- Providing business recommendations for customer segmentation, marketing, fraud detection, and product development

The dashboard includes metrics like total transaction amount, count, state-wise performance, payment type distribution, growth trends, and advanced visualizations (Heatmap, Bubble, Sunburst, Treemap, etc.).

## **Problem Statement**

With the rapid growth of digital payments in India, understanding transaction patterns, state-wise performance, and user behavior is crucial for PhonePe to optimize services, target marketing, detect fraud, and improve product offerings.

- Extract and clean transaction data from PhonePe Pulse
- Perform Exploratory Data Analysis with diverse visualizations

- Build an interactive dashboard for stakeholders

Derived Column: Created Transaction_year_quarter for easier time-based analysis.

Aggregation: Data is already aggregated at State-Year-Quarter-Transaction_type level.

Key Insights Found During Wrangling:
